```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\shiko\\Desktop\\datascienceintership\\TASK 2\\Titanic-Dataset2.csv")
df.info()

```

```python
#deletr unused columns
dff=df.drop(columns=[ 'PassengerId', 'Cabin', 'Ticket'])
```

```python
#delete missing values
df_dropped_na = dff.dropna()
df_reset = df_dropped_na.reset_index(drop=True)
df_reset
```

```python
#check if there is any wrong value
wrong_values=df_reset[(df_reset['Age']<0) | (df_reset['Age']>100) | (df_reset['Fare']<0) ]
wrong_values
```

```python
print(df_reset['Age'].value_counts())
print(df_reset['Sex'].value_counts())
print(df_reset['Fare'].value_counts())
print(df_reset['Survived'].value_counts())
```

```python
sns.boxplot(x=df_reset['Age'])
plt.show()
plt.figure(figsize=(10,6))
df_reset['Fare'].hist(bins=30,edgecolor='black')
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.show()
```
